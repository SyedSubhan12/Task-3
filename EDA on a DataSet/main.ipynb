# Data Loading

# Data Cleaning

```python
#checking for null values
df.isnull().sum()

# Transformation
df['Gender'] = df['Gender'].map({'Male':0, 'Female': 1})
df
```

```python
df['Marital Status'] = df['Marital Status'].map({'Married':0, 'Single':1})
df
```

```python
df['Activity'] = df['Activity'].map({'Professional':0, 'Student':1})
df
```

```python
df['Smoker'] = df['Smoker'].map({'Never':0, 'Often':1, 'Socially':2})
df
```

```python
df['Often A S'] = df['Often A S'].map({'Yes':1, 'No':0})
df
```

```python
df.drop_duplicates(inplace=True)
numeric_data = df.select_dtypes(include=['int64', 'float64'])
```

```python
df.to_csv('Cuisine_rating.csv')
```

# Visualization

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(8, 6))
sns.histplot(df['Overall Rating'], bins=10, kde=True, color='skyblue')
plt.title('Distribution of Overall Rating')
plt.xlabel('Overall Rating')
plt.ylabel('Frequency')
plt.show()
```

# Boxplot of food rating by Gender


```python
plt.figure(figsize=(8, 6))
sns.boxplot(x='Gender', y='Food Rating', data=df, palette='Set2')
plt.title("Food Rating by Gender")
plt.xlabel('Gender')
plt.ylabel('Food Rating')
plt.show()
```

# Countplot of Budget Distribution

```python
plt.figure(figsize=(8, 6))
sns.countplot(x='Budget', data=df,palette='viridis')
plt.title("Distribution of Budget")
plt.xlabel('Budget')
plt.ylabel('Count')
plt.show()
```

# Correlation Matrix Heatmap
